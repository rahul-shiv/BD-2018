```python
import pandas as pd
```

```python
def make_reg(a):
    a=a.split()
    reg=a[0][0]+'.*'+a[-1]
    return reg
```

```python
batsman_clus = pd.read_csv('batsman_cluster.csv')
ipl = pd.read_csv('new.csv',index_col='key')
name_map = dict()
```

```python
#makes a dictionary name_map to map short names to long forms as regex didn't work consistently for this use case
for name in batsman_clus.player_name:
    k = '^'+make_reg(name)+','
    k2 = name+','
    found = ipl.filter(regex = k, axis = 0)
    if len(found)==0:
        found1 = ipl.filter(regex = k2, axis = 0)
        if len(found1)==0:
            k3 = '.*'+make_reg(name)+','
            found2 = ipl.filter(regex = k3, axis = 0)
            if len(found2)==0:
                print(name)
            else:
                mapped = found2.iloc[0].name.split(',')[0]
                name_map[mapped] = name_map.get(mapped,name)
        else:
            mapped = ipl.iloc[0].name.split(',')[0]
            name_map[mapped] = name_map.get(mapped,name)
    else:
        mapped = found.iloc[0].name.split(',')[0]
        name_map[mapped] = name_map.get(mapped,name)
```

```python
bowler_clus = pd.read_csv('bowler_cluster.csv')
```

```python
#adding bowlers to player dictionary
for name in  bowler_clus.player_name:
    k = ', '+make_reg(name)
    k2 = ', '+name
    found = ipl.filter(regex = k, axis = 0)
    if len(found)==0:
        found1 = ipl.filter(regex = k2, axis = 0)
        if len(found1)==0:
            k3 = ', .*'+make_reg(name)
            found2 = ipl.filter(regex = k3, axis = 0)
            if len(found2)==0:
                print(name)
            else:
                mapped = found2.iloc[0].name.split(',')[1]
                name_map[mapped] = name_map.get(mapped,name)
        else:
            mapped = found1.iloc[0].name.split(',')[1]
            name_map[mapped] = name_map.get(mapped,name)
    else:
        mapped = found.iloc[0].name.split(',')[1]
        name_map[mapped] = name_map.get(mapped,name)
```

```python
import json
```

```python
with open('player_map.json','w') as f:
    json.dump(name_map,f)
```
